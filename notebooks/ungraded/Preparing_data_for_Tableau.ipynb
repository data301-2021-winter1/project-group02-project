```python
pathRaw = '../data/raw/recent-grads.csv'
rawData = p1.load_raw(pathRaw)
rawData.head()
```

```python
major_cat_list = rawData['Major_category'].unique()
print(len(major_cat_list))
```

```python
major_cat_list = np.sort(major_cat_list)
major_cat_list
```

```python
count = 0
program_num = []
for i in range( len(major_cat_list) ):
    for k in rawData['Major_category']:
        if major_cat_list[i] == k:
            count += 1
    print(major_cat_list[i], "=", count)
    program_num.append(count)
    count = 0
print(program_num)
```
